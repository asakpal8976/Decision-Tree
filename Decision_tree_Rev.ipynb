# Decision Tree Revision 

  - What is Decision Tree ?
    - A decision tree is a flowchart-like structure where decisions are made at each step (node). It can be used for:
    - Classification: Predicts a category (e.g., yes/no, disease/no disease).
    - Regression: Predicts a continuous value (e.g., house prices).
    - The algorithm falls under the category of supervised learning 

 - Components:
    - Root Node: The starting point where the dataset is divided.
    - Decision Nodes: Points where decisions are made based on certain conditions.
    - Leaf Nodes: Final output nodes that give the prediction (class or value).
    - Branches: Connect nodes and represent decisions or outcomes.

 - Importance of Decision Trees in Machine Learning:
    - Decision trees are widely used in machine learning because they are simple to understand, interpret, and visualize. 
    - They help in making decisions by breaking down a dataset into smaller parts based on specific questions or conditions.
    - The final model helps in classification or regression tasks.
    - They are popular in applications like medical diagnosis, loan approval, and even customer segmentation.

# Algorithm Behind Decision Tree

 - ID3 (Iterative Dichotomizer 3)
    - One of the earliest algorithms, it selects the best feature to split the data using information gain.
    - In case of a regression decision tree, ID3 algorithm makes use of standard deviation reduction
    - ID3 algorithm uses information gain and entropy to construct a classification decision tree
 - Core Idea:
    - ID3 uses Information Gain to decide which feature to split on at each node. The feature with the highest information gain (i.e., the one that provides the most reduction in uncertainty) is chosen for the split.


 - C4.5:
    - An improvement on ID3, it can handle both continuous and discrete data and uses entropy to make splits.
    - does not consider the missing values for calculation of entropy and information gain
    - Pruning: Pruning is a crucial feature introduced in C4.5. It reduces the size of the tree after it has been fully built by removing branches that do not contribute significantly to the decision-making process. This helps to avoid overfitting

 - C5.0:
    - improved version of the C4.5 algorithm, developed by Ross Quinlan.
    - widely used in machine learning for classification tasks
    - known for being faster, more memory-efficient, and accurate



 - Advantages 
    - Easy to Understand and Interpret
    - Works with Both Categorical and Numerical Data
    - No Need for Feature Selection

 - Disadvantages
    - Prone to Overfitting
    - Unstable with Small Changes in Data

# Decision Tree Classifn

 - What is Classification?
    - Classification is a supervised learning task where the goal is to predict the categorical label of new instances based on past observations.
    - The output is a discrete value (class label).

 - How Decision Tree Classification Works:
    - Data Splitting: The algorithm splits the dataset into subsets based on the value of input features, aiming to create pure nodes (where all samples belong to the same class).
Each split is determined by a criterion such as Gini impurity, entropy, or information gain.
    - Node Types: Root Node: The topmost node representing the entire dataset.
Decision Nodes: Nodes that represent the features used to split the data.
Leaf Nodes: Terminal nodes that represent the final class labels.
    - Building the Tree: Starting from the root, the algorithm selects the feature that provides the best split (highest information gain or lowest impurity).
This process continues recursively until one of the stopping criteria is met (e.g., all samples at a node belong to the same class, or a maximum tree depth is reached).
    - Making Predictions: To classify a new instance, traverse the tree from the root to a leaf node based on the feature values of the instance. The class label at the leaf node is the predicted label for that instance.

# What is Gini Index ?

 - The Gini index (or Gini coefficient) is a measure used to quantify inequality or impurity in a dataset
 - The Gini index ranges from 0 to 1:
    - 0 means all samples belong to a single class (pure group).
    - 1 means that the classes are distributed equally (impure group).
 - A variable with higher Gini index is selected to create the sub-node

# What is Entropy ?

 - Entropy is the measure of the amount of uncertainty in the data set
 - If the sample is completely homogenous, then Entropy = 0 
 - If the sample is equally divided, then Entropy = 1

# What is Information Gain ?

 - Information Gain is a metric used in decision trees to determine which feature should be used to split the data at each node
 - Information Gain is the difference between the entropy of the original dataset and the weighted entropy after the dataset is split by a feature.


# Greedy Algorithm 

 - A greedy algorithm is any algorithm that follows the problem-solving heuristic of making the locally optimal choice at each stage.
 - At each step, the algorithm picks the best option available, even if it doesn't guarantee the overall best result.
 - Once a choice is made, the algorithm moves forward and doesn't reconsider previous decisions.
 - Itâ€™s called greedy because at each step, the algorithm "greedily" chooses the best immediate option without worrying about the future

# Decision Tree Regressor

 - Regression is another supervised learning task where the goal is to predict a continuous numeric value based on input features. The output is a real number.

 - How Decision Tree Regression Works:
    - Data Splitting: Similar to classification, the algorithm splits the dataset into subsets based on feature values. However, instead of aiming for pure nodes, it tries to minimize the variance (or mean squared error) within each node.
    - Node Types: The structure is similar to classification trees, with root nodes, decision nodes, and leaf nodes.
Each leaf node in a regression tree represents the average (or median) of the target values in that subset.

    - Building the Tree: Starting from the root, the algorithm selects the feature that results in the lowest mean squared error when splitting the data.The process continues until a stopping criterion is reached (e.g., the maximum depth of the tree or a minimum number of samples per leaf).

    - Making Predictions: To predict the value for a new instance, traverse the tree to find the appropriate leaf node, and the predicted value is the average of the target values in that node.
 - Example: If you're predicting the price of a house based on features like size, location, and number of bedrooms, a regression tree would split the data based on these features and predict a continuous price value at the leaf nodes.

# Standard Deviation Reduction

 - Standard Deviation Reduction (SDR) is one method used for splitting in Decision Tree Regressors.
 - the goal is to find splits that result in groups with less variability in the target variable, and standard deviation is a way to measure that variability.
 - Standard Deviation measures how spread out the values are around the mean.
 - In decision tree regression, the algorithm attempts to split the data in such a way that the resulting groups (after the split) have a lower standard deviation than the original group.
 - The split that results in the greatest reduction in standard deviation is chosen.

# CHAID (Chi Square Automatic Interaction Detector)

 - Chi-Square Automatic Interaction Detector (CHAID) is a decision tree algorithm used primarily for categorical data to identify relationships between a target variable and predictor variables
 - It can handle multi-class target variable 
 - It can be used for predicting both categorical and continuous values 
 - Works effectively with large sample sizes

# Hyper Parameter Tuning

 - Hyperparameter tuning is the process of selecting the best settings (hyperparameters) for a machine learning model to improve its performance.
    - max_depth (Maximum Depth) : The maximum number of levels (or layers) in a decision tree.
    - min_samples_split : The minimum number of samples required in a node to make a split (create child nodes).
    - min_samples_leaf : The minimum number of samples required to be in a leaf node (the final node after a split).
    - max_features : The number of features (input variables) the model considers when deciding the best split at each node.

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
import warnings 
warnings.filterwarnings('ignore')

# Set up the parameter grid for tuning
param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}

# Set up Grid Search
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
grid_search.fit(X_train, Y_train)

# Output the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, Y_test)
print(f"Test set score of the best model: {test_score:.2f}")
```
